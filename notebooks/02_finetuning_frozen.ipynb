```python
from ultralytics import YOLO
import torch

BESTMODEL_PATH = "C:/Users/angel/Desktop/Proyecto/vehicle-damage-assesment/notebooks/runs/mlflow/945788360101712659/b3e1047f83494696bef56b0829607fbc/artifacts/weights/best.pt"

PARTS_YAML = "C:/Users/angel/Desktop/Proyecto/vehicle-damage-assesment/data/processed/parts_dataset/dataset.yaml"

model = YOLO(BESTMODEL_PATH)


def freeze_layer_backbone(model, freeze_count=10):
    print ("Freezing first", freeze_count, "layers of the backbone")
    frozen_layers = 0
    for k, v in model.model.named_parameters():
        try:
            # Extraemos el nÃºmero de capa del nombre
            layer_idx = int(k.split('.')[1]) 
            if layer_idx < freeze_count:
                v.requires_grad = False  # <--- AQUÃ BLOQUEAMOS EL APRENDIZAJE
                frozen_layers += 1
        except:
            pass

freeze_layer_backbone(model, freeze_count=10)

results = model.train(
    data=PARTS_YAML,
    epochs=20,
    imgsz=832,
    batch=2,

    project="Lab_Vehicle_Parts",
    name="Exp4_SGD_Slow_IMG832_Freeze10",

    optimizer="SGD",
    lr0=0.0005,
    lrf=0.0005,

    mosaic = 0.5,
    degrees = 0,

    device = "0",
    workers = 2,
    exist_ok=True

)
```

```python

model = YOLO(BESTMODEL_PATH)
print("\nðŸ“Š 1. Evaluando en modo STANDARD (Sin trucos)...")
# Usamos imgsz=832 porque asÃ­ fue entrenado tu mejor modelo
metrics_std = model.val(split='test', imgsz=832, batch=2, verbose=False)

print("\nðŸš€ 2. Evaluando CON TTA (Test Time Augmentation)...")
print("   (Paciencia: TardarÃ¡ x3 o x4 veces mÃ¡s porque analiza cada foto varias veces)")
# 'augment=True' activa el promediado automÃ¡tico
metrics_tta = model.val(split='test', imgsz=832, batch=2, augment=True, verbose=False)


# --- RESULTADOS ---
map_std = metrics_std.seg.map50
map_tta = metrics_tta.seg.map50
diff = map_tta - map_std

print("\n" + "="*40)
print("ðŸ† RESULTADOS FINALES")
print("="*40)
print(f"ðŸ”¹ mAP50 Standard: {map_std:.4f}")
print(f"ðŸ”¹ mAP50 con TTA:  {map_tta:.4f}")
print(f"ðŸ“ˆ Mejora Real:    {diff:+.4f} ({diff*100:+.2f}%)")
```
