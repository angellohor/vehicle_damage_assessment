```python
from ultralytics import YOLO

dataset_path = "C:/Users/angel/Desktop/Proyecto/vehicle-damage-assesment/data/processed/damages_dataset/data.yaml"

model = YOLO("yolov8n-seg.pt")

training_params = {
    "optimizer": "SGD",
    "lr0": 0.01,
    "lrf": 0.01,

    "scale": 0.6,

    "hsv_h": 0.015,
    "hsv_s": 0.7,
    "hsv_v": 0.4,

    "degrees": 15.0,
    "translate": 0.1,
    "fliplr": 0.5,
    "mosaic": 1.0,
    "mixup": 0.1,
}

results = model.train(
    data=dataset_path,
    epochs=50,
    batch=8,
    imgsz=640,

    project="runs/segmentation/vehicle_damage_assessment",
    name="Exp1_Dmg_HeavyAug",

    device=0,
    **training_params
)
```
